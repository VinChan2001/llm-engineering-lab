```python
system_prompt = """
You analyze the main content of a website and give a short summary.

Instructions:
- Write the entire response in Telugu, but use only English letters.
- Do NOT use Telugu script.
- Example style: "Idi chala clean ga explain chesaru... overall point enti ante..."
- Tone should feel like Mahesh Babu dialogue style: calm, confident, classy, sharp, witty, and smooth.
- Keep it natural and easy to read, like spoken Telugu written in English.
- Do not make it overly dramatic or too filmy.
- Add light punch and humor where it feels natural.
- Ignore navigation menus, headers, footers, buttons, and unrelated website text.
- Focus only on the core content of the page.
- Keep the summary short, crisp, and impactful.

Output requirements:
- Return only markdown.
- Do not wrap the response in a code block.
"""

user_prompt_prefix = """ 
Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""

#the difference between system and user prompt is that system prompt is used to guide the model to
#  behave in a certain way and user prompt is used to actually get the model to do the task

```

```python
messages = [
    {"role": "system", "content": "you are iron man, tony stark"},
    {"role": "user", "content": "hey mr stark, do you like things handed to you?"}
]
response = openai.chat.completions.create(
    model="gpt-4o-mini", messages=messages
)

msg = response.choices[0].message.content
display(Markdown(msg))
```

```python
# See how this function creates exactly the format above

def messages_for(website_contents):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt_prefix + website_contents}
    ]
```

```python
messages_for(vin)
```

```python
# And now: call the OpenAI API. You will get very familiar with this!

def summarize(url):
    website_contents = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = messages_for(website_contents)
    )
    return response.choices[0].message.content
```

```python
summary = summarize("https://www.vinaychanamallu.com/")
display(Markdown(summary))
```
